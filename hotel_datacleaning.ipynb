```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sc
import re
```

```python
dataset=pd.read_csv(r"D:\Data\Internship\Final_hotel.csv")
```

```python
dataset.head(10)
```

```python
dataset.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
dataset["price"]= dataset["price"].str.replace('[#,@,&,$,Rs]', '')
```

```python
dataset['address'] = dataset['address'].str.replace("['',(,),.....,-]", '')
```

```python
dataset["rating"].str.split('\n',expand=True)
```

```python
dataset.drop("no of reviews",axis=1,inplace=True)
```

```python
dataset[["Review","Rating"]] = dataset["rating"].str.extract('(.*)\s+(\d+\.\d+)')
dataset.drop("rating",axis=1,inplace=True)
```

```python
dataset[["distance to city centre(km)","distance to HYD-Rajiv Gandhi Intl(km)"]]=dataset["location"].str.split(r"\n",expand=True)
```

```python
dataset.drop('location',axis=1,inplace=True)
```

```python
dataset["distance to city centre(km)"]=dataset["distance to city centre(km)"].str[0:3]
```

```python
dataset["distance to HYD-Rajiv Gandhi Intl(km)"]=dataset["distance to HYD-Rajiv Gandhi Intl(km)"].str[0:3]
```

```python
dataset["Check_in_date"]=pd.to_datetime(dataset["Check_in_date"])
```

```python
dataset["Check_out_date"]=pd.to_datetime(dataset["Check_out_date"])
```

```python
for amenity in ['Pool','Free parking','Airport transfer','Spa','Restaurant','Gym','Bar','Bathtub','Meeting Facilities','Connecting rooms available','Pet-friendly','Kitchen','Internet access']:
    dataset[amenity] = np.where(dataset['amentities'].str.contains(amenity),'yes','no')
```

```python
dataset.drop('amentities',axis=1,inplace=True)
```
